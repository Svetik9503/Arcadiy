Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска (год выпуска берём из столбца title). Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years;
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска;
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.

```python
#создаем лист с годами 1950 - 2011 через функцию range и метод tolist
years = pd.date_range(start='1950', end='2011', freq='YE').year.tolist()
```

```python
years
```

```python
#пишу функцию, она будет перебирать все значения списка. и смотреть, есть ли в названии год. 
#если включает то пишем его в новый столбец, если не включает, то пишем в этот столбец дефолт знач 1900

#беру значение из столбца title преобразую в тип строку, и ищу в ней значения из списка years 
def production_year(row):
    title = str(row['title'])
    for year in years:
        if str(year) in title:
            return year
    return 1900  

data['year'] = data.apply(production_year, axis=1)
data.head(15)
```

```python
data.head(30)
```

```python
# Удаляем столбец лишний - чтобы заново попробовать 
data.drop( 'rating_by_year', axis=1, inplace=True)
```

### Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга. ###

- Создать новый дф
- сделать группировку по годам. объединяем фильмы, выпущенные в одном году. 
- ищем средний значение рейтинга для фильмов в отдельно взятом году
- сортируем по убыванию значение рейтинга
- переводим в датафрейм 
- Готово! усредненный рейтинг фильмов по годам.

```python
result = (data.groupby('year')['rating'].mean().sort_values(ascending=False).reset_index())
result.head(10)
```

- выводим 10 верхних и 10 нижних строк
- сравнить их средние значения

```python
result.tail(10)
```

#### Утверждение: раньше снимали настоящее кино, не то что сейчас. C ростом года выпуска фильма средний рейтинг становится ниже. 

```python
top_best = result.iloc[0:10, 0:1].mean()
top_best
```

```python
top_worst = result.tail(10).iloc[:, 0:1].mean()
top_worst
```

### ВЫВОД:

```python
if str(top_best) < str(top_worst):
    print('С ростом года выпуска средний рейтинг фильма действительно становится ниже')
else : 
    print('С ростом года выпуска средний рейтинг фильма существенно не меняется')
```

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:

● код должен быть грамотно декомпозирован: каждая функция отвечает за свою конкретную задачу, дублирующиеся функции переиспользуются, а их код не повторяется;
● в коде отсутствуют глобальные переменные, за исключением documents и directories;
● пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу вводом команды «q».

```python
#Исходные данные имеют следующую структуру:

#1. перечень всех документов*

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#2. перечень полок, на которых хранятся документы (если документ есть в documents, он обязательно должен быть и в directories)*

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

# **Задание 1**
Пользователь по команде «p» может узнать владельца документа по его номеру.
Пример работы программы:

Введите команду:
p

Введите номер документа:
10006

Результат:

Владелец документа: Аристарх Павлов

```python
## У меня получилось!!

def search_doc(number_doc):
  for doc in documents:
    if doc['number'] == number_doc:
      return doc ['name']
  print('Документ не найден')

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

while True:
  command = input("Введите команду: ")
  if command.lower() == "q":
    break
  elif command.lower() == "p":
    number_doc = input("Введите номер документа: ");
    result = search_doc(number_doc)
    print('Владелец документа:', result)
```

# **Задание 2 (не обязательное)**
Секретарь по команде «s» может по номеру документа узнать, на какой полке документ хранится.

Пример работы программы:
Введите команду:
s

Введите номер документа:
10006

Результат:
Документ хранится на полке: 2

```python

#У меня получилось!!

# создаю функцию для поиска владельца документа по номеру документа
# внутри нее ищу где ключ будет равен вводу пользователя.
#и прошу вывести из этого словаря значение по ключу Name

def search_doc(number_doc):
  for doc in documents:
    if doc['number'] == number_doc:
      return doc ['name']
  print('Документ не найден')

# создаю функцию для поиска полки где находится документ по номеру документа
# внутри нее ищу значение, которое будет соответствовать вводу пользователя
# точнее, будет его содержать. и прошу вывести ключ от этого значения.
# это номер полки

def search_sh(number_doc):
  for num_sh, docs_list in directories.items():
    if number_doc in docs_list:
      return num_sh
  print('Документ не найден')

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# основной цикл для ввода команд и вызова функций. не останавливается,
# пока пользователь не введет команду 'q'
# цикл просит ввести команду и в зависимости от команды вызывает функцию

while True:
  command = input("Введите команду: ")
  if command.lower() == "q":
    break
  elif command.lower() == "p":
    number_doc = input("Введите номер документа: ");
    result = search_doc(number_doc)
    print('Владелец документа:', result)
  elif command.lower() == "s":
    number_doc = input("Введите номер документа: ");
    result = search_sh(number_doc)
    print('Документ хранится на полке:', result)
```

## Задание 2 (необязательное) ##
Напишите функцию date_range, которая возвращает   список дат за период от start_date до end_date.   Даты должны вводиться в формате yyyy-mm-dd.   В случае неверного формата или при start_date > end_date должен возвращаться пустой список. (date_range(start_date, end_date))

```python
#удалила все разрозненные блоки, все соединила в одном и оставила его
from datetime import datetime, timedelta #импортирую сразу и дататайм и таймдельта

#пишу функцию с проверкой, задаю названия переменных 
def date_range(start_date: str, end_date: str): 
    date_list = list() #инициализирую пустой список

    # тут блок проверки на ошибку ValueError и одновременно перевожу в объект дататайм
    try:
        start_date_datetime = datetime.strptime (start_date, '%Y-%m-%d')
        end_date_datetime = datetime.strptime (end_date, '%Y-%m-%d')

    except ValueError: 
        return date_list

        #чтобы можно было сранивать, вычитать и тд и тп
    if start_date_datetime > end_date_datetime :  #если дата старта больше чем дата окончания, то ошибка
        return date_list

    else : 
        temporary_dt = start_date_datetime  #а иначе можем наполнять список, заводим новую переменную
        while temporary_dt <= end_date_datetime :   #сравниваем
             #добавляем к списку каждое значение переменной темпорари, но перед этим
             #для добавления переводим его из типа дейттайм в строку.
            date_list.append(temporary_dt.strftime('%Y-%m-%d'))  
            temporary_dt += timedelta(days=1)

        print('Список дат:', date_list)
        return date_list

if __name__ == '__main__':
     #тут проверка. подставляем в функцию значения, и сравниваем с тем, что должно получиться
    assert date_range('2022-01-01', '2022-01-03') == ['2022-01-01', '2022-01-02', '2022-01-03']
    assert date_range('2022-01-03', '2022-01-01') == []
    assert date_range('2022-02-30', '2022-02-31') == []

    print('Верное решение')
```

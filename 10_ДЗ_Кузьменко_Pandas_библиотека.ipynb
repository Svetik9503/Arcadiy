```python
Примерный план - 
# отобрать интересующие страны: 
    Латвия, Литва и Эстония
# убрать 
отрицательные значения quantity
# отобрать только нужные категории: 
категории: 4, 12 и 21
# по столбцу год выбрать период 
период с 2005 по 2010 год
# все суммировать 
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики 
(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. 
Не учитывайте в расчётах отрицательные значения quantity.
```

```python
# импортировать пандас
import pandas as pd
```

```python
# импорт данных из файла в датафрейм
data_2 = pd.read_csv('power.csv')
```

```python
# посмотреть, как выглядят данные
data_2.head(30)
```

```python
# основная информация, сведения
data_2.info()
```

```python
# мне только шапку
data_2.head(1)
```

```python
# какие там страны есть в этом списке. найдем уникальные значения
data_2['country'].unique()
```

```python
# Фильтр. мне нужны страны: Латвия, Литва и Эстония
# найдем как называются в этом датафрейме
# фильтр на подстроку - смотрим все страны, содержащие в названии 'tv' и только уникальные значения.
data_2[ data_2['country'].str.contains('tv', case=False) ]['country'].unique()
```

```python
# содержащие в названии 'lit' и только уникальные значения.
data_2[ data_2['country'].str.contains('lit', case=False) ]['country'].unique()
```

```python
# содержащие в названии 'ston' и только уникальные значения.
data_2[ data_2['country'].str.contains('ston', case=False) ]['country'].unique()
```

```python
# фильтруем только страны
# Фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# () | (() | () & ())
filtered_countries = data_2[ (data_2['country']=='Latvia') | (data_2['country']=='Lithuania') | (data_2['country']=='Estonia')]
filtered_countries.head(30)
```

```python
filtered_countries['country'].unique()
```

```python
# количество строк
len(filtered_countries)
```

```python
# фильтры по номеру строки  из data_2
data_2.loc[16000:16005]
```

```python
# фильтруем категории: 4, 12 и 21

filtered_countries = filtered_countries[ (filtered_countries['category']== 4) | (filtered_countries['category']== 12) | (filtered_countries['category']== 21)]
filtered_countries.head(30)
```

```python
# по столбцу год выбрать период период с 2005 по 2010 год

filtered_countries = filtered_countries[ (filtered_countries['year']> 2004) & (filtered_countries['year']< 2011) ]
filtered_countries.head(30)
```

```python
# убрать отрицательные значения quantity

filtered_countries = filtered_countries[ (filtered_countries['quantity'] > 0) ]
filtered_countries.head(30)
```

```python
# посчитаем суммарное потребление

filtered_countries['quantity'].sum()
```
